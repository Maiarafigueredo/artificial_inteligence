#### (b) Demonstrate how Dijkstra’s algorithm would find the shortest path to the solution in Fig.1through diagrams and written explanation. 

Dijkstra Algorithm was published in 1959 by a scientist Edsger Djkstra for fiding the shortest path form one start point to an end point on a graph. 

This algorithm is used to find the shortest path on a weighted graph that can be direct or indirect. The way it works is, the algorithm generates a tree of shortest path from a start node to the other nodes in the graph or else find the shortest path from a start node to a given end node.

Supposed an individual on a game board wants to go from point A to J using the shortest path among connected nodes on the graph. Dijkstra algorithm can be used to achieve this goal.





Algorithm implementation:

The way this algorithm works is, it uses a data structures for querying the solution as it goes along traveling through the nodes and stores the result found arranged by distance from the start node.

In the code bellow was used a Heap data structure for representing the priority queue to execute the dijkstra algorithm.

In this analogy of a gaming graph to find the shortest from A to J. First the algorithm takes the distance from the A node to every other node on the graph with infinity, only to imply that the nodes yet to be visited has not been visited. 

Using a priority queue, ordered by distance, with the shortest distance on the top, every entry visited is added to the queue. The algorithm sorts the entries as it goes along. The run time in this case is:




The V represent the number os node and E is the number of edges.

A loop occurs to very visited node which is marked as visited. And continuo this search of visit a node and update the neighbour node visited with the shortest path found until reach the destination.




```python
graph = {
    'a': {'b':5},
    'b': {'a':5, 'c':8, 'd':9, 'e':3},
    'c': {'b':8,'g':2},
    'd': {'b':9, 'f':2},
    'e': {'b':3, 'i':4},
    'f': {'d':2, 'i':3},
    'g': {'c':2, 'h':3, 'i':6},
    'h': {'g':3, 'j':4},
    'i': {'e':4, 'f': 3, 'g':6, 'j':2},
    'j': {'h':4, 'i':2}
    
}
```

```python
"""
Author: Unknown
Date:10/12/20.
Title:'Shortes Path with Dijkstra's Algoritm
Availability:"https://bradfieldcs.com/algos/graphs/dijkstras-algorithm/"
"""

import heapq


def calculate_distances(graph, starting_vertex):
    distances = {vertex: float('infinity') for vertex in graph}
    distances[starting_vertex] = 0

    pq = [(0, starting_vertex)]
    while len(pq) > 0:
        current_distance, current_vertex = heapq.heappop(pq)

        # Nodes can get added to the priority queue multiple times. We only
        # process a vertex the first time we remove it from the priority queue.
        if current_distance > distances[current_vertex]:
            continue

        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight

            # Only consider this new path if it's better than any path we've
            # already found.
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
    print('Path found: {}'.format(distances))
    return 
```

```python
print(calculate_distances(graph, 'j'))
```

#### Compare the performance of Dijkstra’s algorithm and the A* search algorithm, making reference to heuristics, to find the shortest path to the problem by implementing both algorithms programmatically and comparing the solutions generated in Mark-up (Noteadditional marks for including time statistics) 

In artificial intelligence there’s a need to build systems that would work efficiently so satisfied given constraint, so that A* Search algorithm come to play, it is a path search algorithm. Due to its ultimate efficiency because it has all the functionality of a Dijkstra Algorithm and in addiction it uses heuristics estimates to find the best solution for the problem.

The goal of this informed algorithm is to find the shortest path with smallest cost from a start to end node on weighed graph. 

By using heuristics, in order to gain optimality and completeness, it searches through the graph trying to find the most intelligent way to achieve the given end, by “guessing”, which would be the heuristic, what is the best way to go. It calculates the least distance between the connected nodes and it prioritizes nodes with the lower weight.

In the A* we have two measures for distances, which are the distance to the node and the remaining distance to the end. It uses the following formula. 
 





Where n mean the following node, g is the cost of the weight from on node to another and h is the heurist path from the current node to the goal and f is a sum of g and h. 

So in the case of the game graph, what the A* algorithm does is in every iteration is calculate the cost f(n) and it chose the node according to the least value of f and decide which path to go. 

In artificial intelligence, when dealing with much bigger problem, A* informed search algorithm is very efficient when dealing with real world scenario. 




```python
"""
Author:Vladmir Batacanin
Date:11/12/20.
Title:'Basic AI Concepts: A* Search Algorithm'
Availability: https://stackabuse.com/basic-ai-concepts-a-search-algorithm/
"""

from collections import deque

class Graph:
 

    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list[v]

    # heuristic function with equal values for all nodes
    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'E': 1,
            'D': 1,
            'F': 1,
            'G': 1,
            'H': 1,
            'I': 1,
            'J': 1
            
        }

        return H[n]

    def a_star_algorithm(self, start_node, stop_node):
        # open_list is a list of nodes which have been visited, but who's neighbors
        # haven't all been inspected, starts off with the start node
        # closed_list is a list of nodes which have been visited
        # and who's neighbors have been inspected
        open_list = set([start_node])
        closed_list = set([])

        # g contains current distances from start_node to all other nodes
        # the default value (if it's not found in the map) is +infinity
        g = {}

        g[start_node] = 0

        # parents contains an adjacency map of all nodes
        parents = {}
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None

            # find a node with the lowest value of f() - evaluation function
            for v in open_list:
                if n == None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v;

            if n == None:
                print('Path does not exist!')
                return None

            # if the current node is the stop_node
            # then we begin reconstructin the path from it to the start_node
            if n == stop_node:
                reconst_path = []

                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]

                reconst_path.append(start_node)

                reconst_path.reverse()

                print('Path found: {}'.format(reconst_path))
                return reconst_path

            # for all neighbors of the current node do
            for (m, weight) in self.get_neighbors(n):
                # if the current node isn't in both open_list and closed_list
                # add it to open_list and note n as it's parent
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight

                # otherwise, check if it's quicker to first visit n, then m
                # and if it is, update parent data and g data
                # and if the node was in the closed_list, move it to open_list
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            # remove n from the open_list, and add it to closed_list
            # because all of his neighbors were inspected
            open_list.remove(n)
            closed_list.add(n)

        print('Path does not exist!')
        return None
```

```python
adjacency_list = {
    'A': [('B', 5)],
    'B': [('A', 5), ('C', 8), ('D', 9), ('E', 3)],
    'C': [('B', 8), ('G', 2)],
    'D': [('B', 9), ('F', 2)],
    'E': [('B', 3), ('I', 4)],
    'F': [('D', 2), ('I', 3)],
    'G': [('C', 2), ('H', 3), ('I', 6)],
    'H': [('G', 3), ('J', 4)],
    'I': [('E', 4), ('F', 3), ('G', 6), ('J', 2)],
    'J': [('H', 4), ('I', 2)]
}
graph1 = Graph(adjacency_list)
graph1.a_star_algorithm('A', 'J')
```
